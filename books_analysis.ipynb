```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
X = df1['description']  # Input feature
y = df1['primary_genre']  # Target label
```

```python
vectorizer = TfidfVectorizer(max_features=5000)  
X_transformed = vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Naive Bayes": MultinomialNB()
}
```

```python
accuracy = {}
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy[model_name] = accuracy_score(y_test, y_pred)
print(f"Accuracies for Models:")

for model_name, accuracy_score in accuracy.items():
    print(f"{model_name}: {accuracy_score:.4f}")

```
